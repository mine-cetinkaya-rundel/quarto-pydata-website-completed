## Data

The dataset we'll use comes from the palmerpenguins package.

```python
#| label: import-load-data

from palmerpenguins import load_penguins

penguins = load_penguins()
```

This dataset contains size measurements, clutch observations, and blood isotope ratios for 344 adult foraging Ad√©lie, Chinstrap, and Gentoo penguins observed on islands in the Palmer Archipelago near Palmer Station, Antarctica.[@gorman2014ecological]

# Exploratory analysis

## Body mass vs. flipper length

@fig-penguins shows the relationship between body mass and flipper length of penguins.

```python
#| label: fig-penguins
#| fig-cap: Body mass vs. flipper length of penguins
#| fig-alt: Body mass vs. flipper length of three species of penguins (Adelie, Gentoo, and Chinstrap) showing a positive relationship between the two variables for each species.

# from https://github.com/mcnakhaee/palmerpenguins
g = sns.lmplot(x="flipper_length_mm",
               y="body_mass_g",
               hue="species",
               height=7,
               data=penguins,
               palette=['#FF8C00','#159090','#A034F0'])
g.set_xlabels('Flipper Length (mm)')
g.set_ylabels('Body Mass (g)')
```
